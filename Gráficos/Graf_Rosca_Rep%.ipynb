```python
import matplotlib.pyplot as plt
import numpy as np

# Obtendo os dados do Power BI
categoria = dataset['Linha_Medtronic']
valores = dataset['Vendas Total %']

# Criando o gráfico de rosca
fig, ax = plt.subplots()

# Definindo as cores das categorias
cores = ['#36003e', '#702254', '#862167']
explode = (0, 0.1, 0.1)  # Destacar a segunda e terceira categoria

# Plotando o gráfico de rosca
wedges, texts = ax.pie(
    valores,
    startangle=50,
    explode=explode,
    counterclock=False,
    colors=cores,
    wedgeprops=dict(width=0.5)
)

# Criando um círculo branco no centro para transformar em gráfico de rosca
my_circle = plt.Circle((0, 0), 0.7, color='white')
ax.add_artist(my_circle)

# Garantir que o gráfico seja circular
ax.axis('equal')

# Definindo as propriedades do rótulo e seta
bbox_props = dict(boxstyle="square,pad=0.2", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"), zorder=0, va="center")

# Adicionando rótulos personalizados com setas
for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1) / 2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    percentual = valores[i] / sum(valores) * 100
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = f"angle,angleA=0,angleB={ang}"
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    # Verifica se a categoria é "TERAPIAS ABLATIVAS"
    if categoria[i] == "TERAPIAS ABLATIVAS":
        # Ajusta a posição do rótulo para mais à esquerda e para cima
        ax.annotate(f"{categoria[i]} ({percentual:.1f}%)", xy=(x, y), xytext=(0.5 * np.sign(x), 1.4 * y),
                    horizontalalignment=horizontalalignment, fontsize=11, **kw)
    elif categoria[i] == "ENDOSCOPIA":
        ax.annotate(f"{categoria[i]} ({percentual:.1f}%)", xy=(x,y), xytext=(0.8 * np.sign(x), 1.2 * y),
                    horizontalalignment=horizontalalignment, fontsize=11, **kw)
    else:
        # Mantém a posição padrão para outras categorias
        ax.annotate(f"{categoria[i]} ({percentual:.1f}%)", xy=(x, y), xytext=(1.0 * np.sign(x), 1.2 * y),
                    horizontalalignment=horizontalalignment, fontsize=11, **kw)

# Exibir o gráfico
plt.show()
```
